```python
from sklearn.datasets import load_wine
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=load_wine()
dataset=pd.DataFrame(df.data)
dataset
```

```python
dataset.columns=df.feature_names
dataset
```

```python
df.target.shape
```

```python
dataset['Quality']=df.target
dataset.head()
```

```python
dataset.corr()
```

```python
x=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
```

```python
lin_reg=LinearRegression()
mse=cross_val_score(lin_reg,x,y,scoring='neg_mean_squared_error',cv=5)
mean_mse=np.mean(mse)
print(mean_mse)
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
rid=Ridge()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,50,55,100]}
rid_reg=GridSearchCV(rid,parameters,scoring='neg_mean_squared_error',cv=5)
rid_reg.fit(x,y)
```

```python
print(rid_reg.best_params_)
print(rid_reg.best_score_)
```
